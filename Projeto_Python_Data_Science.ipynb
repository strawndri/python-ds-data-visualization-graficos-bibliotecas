# 1. Conhecendo a biblioteca Matplotlib

## 1.1 Conhecendo os dados

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/strawndri/python-ds-data-visualization-graficos-bibliotecas/main/dados/imigrantes_canada.csv')
df.head()
```

```python
df.info()
```

### Filtrando informações apenas do Brasil

```python
df.set_index('País', inplace = True)
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
brasil = df.loc['Brasil', anos]
```

```python
brasil_dict = {'ano': brasil.index.tolist(),
               'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil.head()
```

## 1.2 Criando o primero gráfico com Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show()
```

## 1.3 Adicionando títulos e rótulos

```python
plt.figure(figsize=(8, 4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show()
```

# 2. Criando figuras com Matplotlib

## 2.1 Criando uma figura

```python
fig, ax = plt.subplots(figsize = (8, 4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

## 2.2 Adicionando títulos e rótulos em figuras

```python
fig, ax = plt.subplots(figsize = (8, 4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

## 2.3 Crindo subplots em uma direção

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da imigraçaõ do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid()

plt.show()
```

```python
dados_brasil.describe()
```

## 2.4 Criando subplots em duas direções

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace = 0.5, wspace = 0.3)
fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá de 1980 a 2013')

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
  ax.set_xlabel('Ano')
  ax.set_ylabel('Número de Imigrantes')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

# 3. Customizando com Matplotlib

## 3.1 Personalizando visualizações com Matplotlib

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3)
ax.set_title('Imigração do Brasil para o Canadá (1980 a 2013)', fontsize=16, loc='left')
ax.set_xlabel('Ano', fontsize=12)
ax.set_ylabel('Número de imigrantes', fontsize=12)
ax.xaxis.set_tick_params(labelsize=11)
ax.yaxis.set_tick_params(labelsize=11)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.grid(linestyle=':')
plt.show()
```

## 3.2 Alterando as cores

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3, color = '#9cd33b')
ax.set_title('Imigração do Brasil para o Canadá (1980 a 2013)', fontsize=16, loc='left')
ax.set_xlabel('Ano', fontsize=12)
ax.set_ylabel('Número de imigrantes', fontsize=12)
ax.xaxis.set_tick_params(labelsize=11)
ax.yaxis.set_tick_params(labelsize=11)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.grid(linestyle=':')
plt.show()
```

```python
df.head()
```

```python
america_sul = df.query('Região == "América do Sul"')
america_sul
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid',
         'purple', 'brown', 'slateblue', 'gray', 'olive',
         'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul.index, america_sul['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá (1980 a 2013)', loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=12)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=11)
ax.yaxis.set_tick_params(labelsize=11)
plt.show()
```

```python
america_sul_sorted = america_sul.sort_values('Total', ascending = True)
```

```python
cores = []

for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('#9cd33b')
  else:
    cores.append('#d6d6d6')

fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013', loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=12)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=11)
ax.yaxis.set_tick_params(labelsize=11)
plt.show()
```

## 3.3 Adicionando anotações

```python
cores = []

for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('#9cd33b')
  else:
    cores.append('#d6d6d6')

fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013', loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=12)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=11)
ax.yaxis.set_tick_params(labelsize=11)

for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v + 200, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

plt.show()
```
